```python
ratings=pd.read_csv("/content/netflix-movie-rating-dataset/Netflix_Dataset_Rating.csv")
ratings
```

```python
# counting unique values of movie_id
n = len(pd.unique(ratings['Movie_ID']))

print("No.of.unique values :", n)
```

```python
movies=pd.read_csv("/content/netflix-movie-rating-dataset/Netflix_Dataset_Movie.csv")
movies
```

#Merge the 2 dataframes

```python
movies_ratings=pd.merge(ratings,movies)
movies_ratings
```

#Some Information about merged dataframe

```python
#computing number of rows
rows = len(movies_ratings.axes[0])
print("Number of Rows: ", rows)

# computing number of columns
cols = len(movies_ratings.axes[1])
print("Number of Columns: ", cols)
```

#How many unique movies we have:  we have 1342 unique movies




```python
# counting unique values of movie_id/Name in merged df (movies_ratings)
unique_movies_names = len(pd.unique(movies_ratings['Name']))
print("No.of.unique movies :", unique_movies_names)
```

#How many unique user id we have in merged dataframe (movies_ratings):  

```python
unique_users = len(pd.unique(movies_ratings['User_ID']))
print("No.of.unique users :", unique_users)
```

#Therefore if we transformed it to pivot table (its index is the users_id & (columns are the unique movies)  ..... the size of the pivot table is (143458 rows × 1342 columns)

#Transform it to pivot table:

```python
movies_ratings=movies_ratings.pivot_table(index=["User_ID"],columns=["Name"],values="Rating")
movies_ratings
```

```python
type(movies_ratings)
```

#Search for all movies contains word (Wild) for example in (movies) dataframe

#I have 2 ways to make this :
#1- using for loop to iterate over movies['Name'] series

```python
wild_movies=[val for val in movies['Name'] if 'Wild' in val ]
print(wild_movies) #It is a list
print("number of movies contains wild word:  ",len(wild_movies))
```

#2-Using -- Series.str.contains("specific word") - which returns true/false depending on occurance of the "specific word" in the values of the series.

https://www.geeksforgeeks.org/python-pandas-series-str-contains/

```python
wild_movies=movies['Name'][movies['Name'].str.contains("Wild")]
print(wild_movies) #It is a dataframe (it shows the index too)
print("number of movies contains wild word:  ",len(wild_movies))
print(type(movies_ratings))
```

#I want to see rating of users for one of the movies .. for example (Wild Things) movie..So, I will check movies_ratings dataframe

```python
wild_Things_ratings=movies_ratings['Wild Things']
wild_Things_ratings.head(10)
```

#LAST STEP:

#I need to check correlation between ratings of this movie (Wild Things) and ratings of other movies

```python
correlation=movies_ratings.corrwith(wild_Things_ratings)
correlation.size
```

```python
Similar_movies=correlation.sort_values(ascending=False)
```
